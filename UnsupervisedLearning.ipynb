Replace 'XLABEL' and 'YLABEL' with the respective labels in your data set.

What is your X and Y label?

This will try to visualize the relationship between a character's Base HP and Base Attack.
X label will be Base HP
Y label will be Base Attack

Import packages

```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
XLABEL = 'Base HP'
YLABEL = 'Base ATK'
```

Import and explore the data set

```python
df = pd.read_csv('cleanedfile.csv')
```

2. Display columns and describe the data set

```python
df.describe()
```

Scatter plot

```python
plt.scatter(df[XLABEL],df[YLABEL])
plt.xlabel(XLABEL)
plt.ylabel(YLABEL)
```

Fit and predict three clusters for the data

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[[XLABEL,YLABEL]])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1[XLABEL],df1[YLABEL],color='green')
plt.scatter(df2[XLABEL],df2[YLABEL],color='red')
plt.scatter(df3[XLABEL],df3[YLABEL],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel(XLABEL)
plt.ylabel(YLABEL)
plt.legend()
```

Fit the model using MinMaxScaler()

```python
scaler = MinMaxScaler()

scaler.fit(df[[YLABEL]])
df[YLABEL] = scaler.transform(df[[YLABEL]])

scaler.fit(df[[XLABEL]])
df[XLABEL] = scaler.transform(df[[XLABEL]])
```

```python
plt.scatter(df[XLABEL],df[YLABEL])
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[[XLABEL,YLABEL]])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1[XLABEL],df1[YLABEL],color='green')
plt.scatter(df2[XLABEL],df2[YLABEL],color='red')
plt.scatter(df3[XLABEL],df3[YLABEL],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.legend()
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[[XLABEL,YLABEL]])
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
